## (1) Data Description
The dataset was collected from a Minecraft research server run by a research group in computer science at UBC. It includes two files: **players.csv**, containing information about each unique player, and **sessions.csv**, containing information about each play session. The two files can be linked through a shared variable, **'hashedEmail'**, which serves as the **main key**. 

```r
players <- read_csv("players.csv")
players
```

```r
sessions <- read_csv("sessions.csv")
sessions
```



## Dataset Structure & Variables
### **players.csv**
 - **Number of observations (rows):** 196
 - **Number of variables (columns):** 7
 - 'experience' *(character)* - player's self-reported skill level.
 - 'subscribe' *(logical)* - TRUE/FALSE indicating whether the player subscribed to the newsletter.
 - 'hashedEmail' *(character)* - hashed version of the player's email.
 - 'played_hours' *(numeric)* - total number of hours the player has spent playing on the server.
 - 'name' *(character)* - name of the player.
 - 'gender' *(character)* - player's gender as reported during registration.
 - 'Age' *(numeric)* - player's age in years. 
### **sessions.csv**
 -  **Number of observations (rows):** 1,535
 -  **Number of variables (columns):** 5
 -  'hashedEmail' *(character)* - hashed email linking to player data.
 -  'start_time' *(character)* - start date and time of the gameplay session.
 -  'end_time' *(character)* - end date and time of the gameplay session.
 -  'original_start_time' *(numeric)* - timestamp (milliseconds) representing session start time.
 -  'original_end_time' *(numeric)* - timestamp (milliseconds) representing session end time.


## Rounding Summary Statistics
All summary statistics were rounded to **two decimal places** using the 'round()' function inside the 'summarise()' command. For each numeric variable (e.g., 'played_hours', 'Age', 'original_start_time', 'original_end_time'), the minimum, mean, and maximum values were calculated and rounded with round(value, 2) to ensure consistent numeric precision across both datasets. 

```r
library(dplyr)
```

```r
played_hours_summary <- players |>
summarise(min = round(min(played_hours, na.rm = TRUE), 2),
          mean = round(mean(played_hours, na.rm = TRUE), 2),
          max = round(max(played_hours, na.rm = TRUE), 2))

played_hours_summary
players_age_summary <- players |>
summarise(min = round(min(Age, na.rm = TRUE), 2),
          mean = round(mean(Age, na.rm = TRUE), 2),
          max = round(max(Age, na.rm = TRUE), 2))
players_age_summary
```

```r
original_start_time_summary <- sessions |>
summarise(min_start = round(min(original_start_time, na.rm = TRUE), 2),
          mean_start = round(mean(original_start_time, na.rm = TRUE), 2),
          max_start = round(max(original_start_time, na.rm = TRUE), 2))

original_start_time_summary
original_end_time_summary <- sessions |>
summarise(min_end = round(min(original_end_time, na.rm = TRUE), 2),
          mean_end = round(mean(original_end_time, na.rm = TRUE), 2),
          max_end = round(max(original_end_time, na.rm = TRUE), 2))
original_end_time_summary
```

## (3) Exploratory Data Analysis and Visualization

I already loaded both 'players.csv' and 'sessions.csv' into R using the 'read_csv()' function from the **tidyverse** package. I also used 'glimpse()' to check the number of rows, columns, and types of variables. I will calculate the mean for each numeric variable in players.csv and round the results to two decimal places. 

```r
players_mean <- players |>
summarise(mean_Age = round(mean(Age, na.rm = TRUE), 2),
          mean_played_hours = round(mean(played_hours, na.rm = TRUE), 2))
players_mean
```

```r
played_hours_vs_age_plot <- players |> 
     ggplot(aes(x = Age, y = played_hours)) +
     geom_point() +
     labs(x = "Player Age (years)", y = "Played Times (Hours)",
     title = "Played Hours vs. Age")
played_hours_vs_age_plot 
```

Insight: Younger players (around 15-25 years old) seem to play more compared to older players. 

```r
played_hours_histogram <- players |>
ggplot(aes(x = played_hours)) +
geom_histogram(bins = 30, fill = "darkblue", color = "white") +
labs(x = "Played Hours", y = "Number of Players", title = "Distribution of Played Hours")

played_hours_histogram
```

Insight: Most players played for a short time (mostly never played), while a few players played many hours. 

```r
played_hours_by_experience_level <- players |> 
group_by(experience) |>
summarise(mean_hours = mean(played_hours, na.rm = TRUE)) |>
ggplot(aes(x = experience, y = mean_hours, fill = experience)) +
geom_bar(stat = "identity") +
labs(x = "Experience Level", y = "Mean Played Hours",
     title = "Average Played Hours by Experience Level") + 
theme_minimal()
played_hours_by_experience_level
```

Insight: Players with the Regular experience level have the highest average played hours. This means that experience level still might be useful for predicting high-usage players, but the results may be influenced by a few Regular players who played many hours. 

```r
played_hours_by_subscription_status <- players |> 
group_by(subscribe) |>
summarise(mean_hours = mean(played_hours, na.rm = TRUE)) |>
ggplot(aes(x = subscribe, y = mean_hours, fill = subscribe)) +
geom_bar(stat = "identity") +
labs(x = "Subcribe to Newsletter", y = "Mean Played Hours",
     title = "Average Played Hours by Subscription Status") + 
theme_minimal()
played_hours_by_subscription_status
```

Insight: Players who subscribed to the newsletter tend to have higher average played hours, meaning subscription status might be a useful predictor for high-usage players.  

## (4) Method and Plan

### Proposed Method 
I plan to use the **K-Nearest Neighbors(KNN) classification** method to predict whether a player will be a **high-usage player**(1 = high usage, 0 = low usage). KNN works by comparing a player to others in the dataset and predicting based on the majority class of the closest neighbors. Since KNN classification is a simple machine learning algorithm, it can make predictions based on similarity, as in players with similar characteristics (age, experience, gender, subscription) are likely to have similar playtime behavior. 

### Assumptions and Limitations
- All features must be **numeric and scaled** properly so that no variable dominates distance calculations.
- Players with similar features are expected to behave similarly. 
- KNN classification can be slow on large datasets since it must compute distances for every new observation.
- It is sensitive to scaling and the choice of k-neighbors.

### Comparison
I will try different values of k-neighbors and compare their accuracy. I will also compare the performance with KNN-regression to see which predicts better (Model performance will be evaluated using **accuracy**, **precision**, and **recall**).
